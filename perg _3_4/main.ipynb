## Trabalhando nos dados

#### Pergunta 3: Quais os 5 estados com o preço médio da gasolina e do etanol mais caros?

```python
# Levando em consideração os 2 meses e os 2 combustíveis juntos,Gasolina e Etanol.
dados_agrupados_estado = conjunto_dados.copy()
media_combs_estados = dados_agrupados_estado.groupby(dados_agrupados_estado['Estado - Sigla'])
print('OS 5 ESTADOS COM O MAIOR PREÇO MÉDIO DE GASOLINA COMUM/ADITIVADA E ETANOL NO BIMESTRE.')
display(media_combs_estados[['Valor de Venda']].mean().sort_values(by = 'Valor de Venda',ascending = False).head())
```

```python
# Fazendo a média bimestral com a Gasolina,Gasolina Aditivada e o Etanol separados.
media_gasolinadit_estado = dados_agrupados_estado.loc[dados_agrupados_estado['Produto'] == 'GASOLINA ADITIVADA'].groupby(dados_agrupados_estado['Estado - Sigla'])
print('OS 5 ESTADOS COM O MAIOR PREÇO MÉDIO DE GASOLINA ADITIVADA NO BIMESTRE.')
df_media_gasolinadt_estado = media_gasolinadit_estado[['Valor de Venda']].mean()
display(df_media_gasolinadt_estado.sort_values(by = 'Valor de Venda',ascending = False).head())
print('\n')
media_gasolina_estado = dados_agrupados_estado.loc[dados_agrupados_estado['Produto'] == 'GASOLINA'].groupby(dados_agrupados_estado['Estado - Sigla'])
print('OS 5 ESTADOS COM O MAIOR PREÇO MÉDIO DE GASOLINA COMUM NO BIMESTRE.')
df_media_gasolina_estado = media_gasolina_estado[['Valor de Venda']].mean()
display(df_media_gasolina_estado.sort_values(by = 'Valor de Venda',ascending = False).head())
print('\n')
print('OS 5 ESTADOS COM O MAIOR PREÇO MÉDIO DE ETANOL NO BIMESTRE.')
media_etanol_estado = dados_agrupados_estado.loc[dados_agrupados_estado['Produto'] == 'ETANOL'].groupby(dados_agrupados_estado['Estado - Sigla'])
df_media_etanol_estado = media_etanol_estado[['Valor de Venda']].mean()
display(df_media_etanol_estado.sort_values(by = 'Valor de Venda',ascending = False).head())
```

#### Pergunta 4: Qual o preço médio da gasolina e do etanol por estado?

```python
# Fazendo a média bimestral da Gasolina,da Gasolina Aditivada e do Etanol.
df_media_comb_estado = pd.concat([df_media_gasolinadt_estado,df_media_gasolina_estado,df_media_etanol_estado],axis = 1)
df_media_comb_estado.columns = ['Gasolina Aditivada','Gasolina Comum','Etanol']
print('Tabela com a média dos combustíveis por estado:\n')
df_media_comb_estado
```
