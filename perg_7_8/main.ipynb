```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

## Tratamento

```python
csv = pd.read_csv('dados/precos-gasolina-etanol-07.csv', delimiter=';')
comb_jul = pd.DataFrame(csv)
comb_jul.head(1)
```

```python
csv = pd.read_csv('dados/precos-gasolina-etanol-08.csv', delimiter=';')
comb_ago = pd.DataFrame(csv)
comb_ago.head(1)
```

```python
conjunto_dados = pd.concat([comb_jul, comb_ago])
conjunto_dados.head(1)
```

```python
conjunto_dados.reset_index(0,drop=True,inplace=True)
```

```python
((conjunto_dados.isnull().sum() / conjunto_dados.shape[0]) * 100).sort_values(ascending=False)
```

```python
conjunto_dados.drop('Valor de Compra', axis=1, inplace=True)
```

```python
conjunto_dados['Valor de Venda'] = conjunto_dados['Valor de Venda'].apply(lambda y: float(str(y.replace(',','.'))))
```

```python
conjunto_dados.Complemento.fillna('0', inplace=True)
```

```python
conjunto_dados.head()
```

```python
conjunto_dados['Data da Coleta'] = pd.to_datetime(conjunto_dados['Data da Coleta'],format='%d/%m/%Y')
conjunto_dados.head(1)
```

## Trabalhando nos dados

```python
# 7. Qual a região que possui o maior valor médio da gasolina?

# Criando um recorte do DF original onde temos apenas os registros da Gasolina
df_gas = conjunto_dados.loc[(conjunto_dados['Produto'] == 'gasolina'.upper()), :]
```

```python
df_gas.groupby(['Regiao - Sigla', 'Produto'])[['Valor de Venda']].mean().sort_values(by='Valor de Venda', ascending=False)
```

```python
# 8. Qual a região que possui o menor valor médio do etanol?

# Criando um recorte do DF original onde temos apenas os registros do Etanol
df_et = conjunto_dados.loc[(conjunto_dados['Produto'] == 'etanol'.upper()), :]
```

```python
df_et.groupby(['Regiao - Sigla', 'Produto'])[['Valor de Venda']].mean().sort_values(by='Valor de Venda', ascending=True)
```
